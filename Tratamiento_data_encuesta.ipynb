# **Análisis de datos faltantes**

Con el objetivo de identificar la presencia de datos faltantes en el conjunto de datos de la encuesta realizada a los alumnos del Instituto Superior Tecnológico Guayaquil, se utilizó un mapa de calor (heatmap) generado a partir de la biblioteca seaborn.

El siguiente código muestra cómo se creó el mapa de calor para visualizar los datos faltantes en las primeras 10 columnas del conjunto de datos:

```python
cols = data.columns[:10] # first 30 columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(data[cols].isnull(), cmap=sns.color_palette(colours))
```

En el resultado obtenido, se observa que todas las celdas de las primeras 10 columnas aparecen de color amarillo, lo que indica que hay datos faltantes en esas columnas. Este patrón de datos faltantes debe ser abordado adecuadamente en el análisis posterior para garantizar la integridad y confiabilidad de los resultados.

Con el objetivo de identificar la presencia de datos faltantes en el conjunto de datos de la encuesta realizada a los alumnos del Instituto Superior Tecnológico Guayaquil, se realizó un análisis del porcentaje de valores faltantes por atributo de entrada.

El siguiente código muestra cómo se calculó y presentó el porcentaje de valores faltantes para cada atributo de entrada:

```python
# Porcentaje de los valores faltantes por atributo de entrada
for col in data.columns:
    pct_missing = np.mean(data[col].isnull())
    print('| {:<20} | {:>6.2f}% |'.format(col, round(pct_missing*100, 2)))

# Imprimir línea de separación
print('-' * 31)
```

En el resultado obtenido, se observa que todos los campos tienen un 0% de datos faltantes. Esto indica que no se encontraron valores faltantes en ninguna de las variables del conjunto de datos de la encuesta.

El siguiente código muestra cómo se realizó esta transformación:

```python
# Obtener lista de nombres de columnas categóricas
columnas_categoricas = data.select_dtypes(include='object').columns
```

```python
# Definir los límites y etiquetas para los rangos de edad
bins = [0, 18, 30, 40, 50, 60, float('inf')]
labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '60+']

# Transformar la columna "edad" a una variable categórica
data['edad_categorica'] = pd.cut(data['Edad'], bins=bins, labels=labels, right=False)
```

En el código anterior, se definen los límites y etiquetas para los rangos de edad. Luego, se utiliza la función pd.cut para asignar cada valor de la columna "edad" a su respectivo rango y se crea una nueva columna llamada "edad_categorica" que contiene la variable categórica resultante.

Posteriormente, se eliminaron las columnas "id" y "edad" de la data, ya que no son relevantes para el análisis y no aportan información significativa:

```python
# Eliminar las columnas "id" y "edad" de la data
data_sin_id_y_edad = data.drop(['id', 'Edad'], axis=1)
```

La transformación de la columna "edad" a una variable categórica y la eliminación de las columnas "id" y "edad" permiten asegurar que los datos estén en el formato adecuado para el análisis posterior."

Se realizó un análisis de las tablas de frecuencia para cada una de las columnas del conjunto de datos. Las tablas de frecuencia proporcionan información sobre la distribución de los valores en cada columna, permitiendo obtener una visión general de la frecuencia de aparición de cada valor y su respectivo porcentaje.

El siguiente código muestra cómo se generaron las tablas de frecuencia para cada columna:

```python
# Generar las tablas de frecuencia para cada columna
tablas_frecuencia = {}
for columna in data_sin_id_y_edad.columns:
    tabla_frecuencia = data_sin_id_y_edad[columna].value_counts().reset_index()
    tabla_frecuencia.columns = ['Valor', 'Frecuencia']
    total = tabla_frecuencia['Frecuencia'].sum()
    tabla_frecuencia['Porcentaje'] = (tabla_frecuencia['Frecuencia'] / total) * 100
    tabla_frecuencia['Porcentaje_decimal'] = (tabla_frecuencia['Frecuencia'] / total) * 100
    tablas_frecuencia[columna] = tabla_frecuencia

    # Convertir la columna 'Valor' a tipo string
    tabla_frecuencia['Valor'] = tabla_frecuencia['Valor'].astype(str)

    # Crear el estilo de la tabla
    estilo_tabla = tabla_frecuencia.style \
        .format({'Porcentaje': '{:.2f}%'.format}) \
        .set_table_styles([
            {'selector': 'th', 'props': [('border', '1px solid black'), ('background-color', 'lightblue')]},
            {'selector': 'td', 'props': [('border', '1px solid black')]}
        ]) \

    # Agregar la fila y columna total
    tabla_frecuencia.loc['Total'] = tabla_frecuencia.sum()
    tabla_frecuencia.at['Total', 'Valor'] = 'Total'
    tabla_frecuencia.at['Total', 'Porcentaje'] = '100%'

    # Imprimir la tabla de frecuencia con estilo
    print('Tabla de Frecuencia para la columna {}'.format(columna))
    display(estilo_tabla)

    print('\n')
```

Se utilizó Python y la biblioteca Matplotlib para generar gráficos que representan las tablas de frecuencia de los datos recopilados. El código utilizado para generar los gráficos se muestra a continuación:

```python
colores_comunes = ['red', 'blue', 'yellow', 'green', 'purple', 'pink']
for columna, tabla_frecuencia in tablas_frecuencia.items():
    plt.figure()
    tabla_frecuencia = tabla_frecuencia.drop('Total', axis=0)
    if columna == 'edad_categorica' or columna == 'Género':
        
        # Obtener el número de categorías
        num_categorias = len(tabla_frecuencia)

        # Asignar colores comunes a las barras
        colores = colores_comunes[:num_categorias]

        plt.bar(tabla_frecuencia['Valor'], tabla_frecuencia['Porcentaje_decimal'], width=0.8, color=colores)
        plt.xlabel('Valor')
        plt.ylabel('Porcentaje')
        plt.title('Gráfico de barras para la columna {}'.format(columna))
        plt.xticks(rotation=90)

        # Agregar etiquetas de porcentaje encima de cada barra
        for i, v in enumerate(tabla_frecuencia['Porcentaje_decimal']):
            plt.text(i, v, '{:.1f}%'.format(v), ha='center', va='bottom')
    else:
        plt.pie(tabla_frecuencia['Porcentaje_decimal'], labels=tabla_frecuencia['Valor'], autopct='%1.1f%%', startangle=90)
        plt.axis('equal')
        plt.title('Gráfico de pastel para la columna {}'.format(columna))

    plt.tight_layout()
    plt.show()
```

El código utiliza el enfoque de gráficos de barras para las columnas 'edad_categorica' y 'Género', mientras que se utiliza el gráfico de pastel para las demás columnas. Para las columnas 'edad_categorica' y 'Género', se asignaron colores comunes a las barras para facilitar la distinción entre las diferentes categorías.

Se agregaron etiquetas de porcentaje encima de cada barra en el gráfico de barras para visualizar claramente los valores correspondientes. En el gráfico de pastel, se incluyeron etiquetas con el porcentaje correspondiente a cada categoría.
